```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('placement.csv',usecols=[1,2,3])

df.head()
```

```python
sns.kdeplot(df['iq'])
```

```python
sns.kdeplot(df['cgpa'])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop('placement',axis=1),df['placement'],test_size=0.3,random_state=0)

X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled,columns=X_train.columns)

X_test_scaled = pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

```python
np.round(X_train.describe(),1)
```

```python
np.round(X_train_scaled.describe(),1)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,figsize=(10,5))

ax1.set_title('Before Scaling')
ax1.scatter(X_train['iq'],X_train['cgpa'] , c=y_train)

ax2.set_title('After Scaling')
ax2.scatter(X_train_scaled['iq'],X_train_scaled['cgpa'] , c=y_train)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,figsize=(10,5))

ax1.set_title('Before Scaling')
sns.kdeplot(X_train['iq'], ax=ax1)

ax2.set_title('After Scaling')
sns.kdeplot(X_train_scaled['iq'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,figsize=(10,5))

ax1.set_title('Before Scaling')
sns.kdeplot(X_train['cgpa'], ax=ax1)

ax2.set_title('After Scaling')
sns.kdeplot(X_train_scaled['cgpa'], ax=ax2)
plt.show()
```
