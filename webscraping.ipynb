```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

final = pd.DataFrame()

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/124.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
    "Referer": "https://www.pakwheels.com/"
}

for j in range(1, 11):
    url = f"https://www.pakwheels.com/used-cars/search/-/?page={j}"

    try:
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f"❌ Failed to fetch page {j}: {e}")
        continue

    soup = BeautifulSoup(response.text, "lxml")

    company = soup.find_all("div", class_="col-md-9 grid-style")

    car_name, car_price, city, model_year, used_km, fuel_type, CC, type_of_vehicle = ([] for _ in range(8))

    for i in company:
        # Car name
        car_name.append(i.find("h3").get_text(strip=True) if i.find("h3") else None)

        # Price
        price_tag = i.find("div", class_="price-details generic-dark-grey")
        car_price.append(price_tag.get_text(strip=True) if price_tag else None)

        # City
        city_tag = i.find("ul", class_="list-unstyled search-vehicle-info fs13")
        city.append(city_tag.get_text(strip=True) if city_tag else None)

        # Vehicle info list
        info_ul = i.find("ul", class_="list-unstyled search-vehicle-info-2 fs13")
        info_items = info_ul.find_all("li") if info_ul else []

        model_year.append(info_items[0].get_text(strip=True) if len(info_items) > 0 else None)
        used_km.append(info_items[1].get_text(strip=True) if len(info_items) > 1 else None)
        fuel_type.append(info_items[2].get_text(strip=True) if len(info_items) > 2 else None)
        CC.append(info_items[3].get_text(strip=True) if len(info_items) > 3 else None)
        type_of_vehicle.append(info_items[4].get_text(strip=True) if len(info_items) > 4 else None)

    df = pd.DataFrame({
        "Car Name": car_name,
        "Car Price": car_price,
        "City": city,
        "Model Year": model_year,
        "Used KM": used_km,
        "Fuel Type": fuel_type,
        "CC": CC,
        "Type of Vehicle": type_of_vehicle
    })

    final = pd.concat([final, df], ignore_index=True)

    print(f"✅ Page {j} scraped, {len(df)} records added.")
    time.sleep(2)  # polite delay to avoid blocking
```
