```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('train.csv',usecols=['GarageQual','FireplaceQu','SalePrice'])
```

```python
df.head()
```

```python
df.isnull().mean()*100
```

```python
df['GarageQual'].value_counts().sort_values(ascending=False).plot.bar()

plt.xlabel('GarageQual')
plt.ylabel('Count')
```

```python
df['GarageQual'].mode()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df[df['GarageQual']=='TA']['SalePrice'].plot(kind='kde',ax=ax,label='TA')

df[df['GarageQual'].isnull()]['SalePrice'].plot(kind='kde',ax=ax,label='TA')

lines,labels = ax.get_legend_handles_labels()
labels = ['Houses with TA','Houses with NA']
ax.legend(lines,labels , loc='best')

plt.title('GarageQuality')

```

```python
temp = df[df['GarageQual']=='TA']['SalePrice']
```

```python
df['GarageQual'].fillna('TA',inplace=True)
```

```python
df['GarageQual'].value_counts().sort_values(ascending=False).plot.bar()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

temp.plot(kind='kde',ax=ax)

df[df['GarageQual']=='TA']['SalePrice'].plot(kind='kde',ax=ax,color='red')

lines,labels = ax.get_legend_handles_labels()
labels = ['Original Value','Imputed Value']
ax.legend(lines,labels , loc='best')

plt.title('GarageQuality')

```

```python
df['FireplaceQu'].value_counts().sort_values(ascending=False).plot.bar()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df[df['FireplaceQu']=='Gd']['SalePrice'].plot(kind='kde',ax=ax,label='Gd')

df[df['FireplaceQu'].isnull()]['SalePrice'].plot(kind='kde',ax=ax,label='Gd')

lines,labels = ax.get_legend_handles_labels()
labels = ['Houses with Gd','Houses with NA']
ax.legend(lines,labels , loc='best')

plt.title('GarageQuality')

```
